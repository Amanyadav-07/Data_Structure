#Q1. Discuss string slicing and provide example.

####String Slicing:- we can return a range of characters by using the slice syntax.Specify the start index and the end index, separated by a colon, to return a part of the string.

1. The first character has index 0.
2. Slice From the Start:- By leaving out the start index, the range will start ast the first charactor
1. Slice To the End:- By leaving out the end index, the range will go to he end.
4. Negative Indexing:- Use Negative Indexes to start the slice from the end of the string.
5. slicing is a technique to extract a substring from a given string by specifying the start and end indices, and optionally a step value.The syntax for string slicing is:-
  
  string[start:stop:step]

  A. Start:- The index of the first character to include in the slice.
  Defaults to 0, which means the beginning of the string.
  
  B. Stop:- the index of the first character to exclude from the slice. Defaults to the length of the string, which means the end of the string.

  C. Step:- The number of characters to skip b/w each charater in the slice. Defaults to 1, which means no skipping.

```python
# Slice From the Start

str = "Hello, World!"
print(str[0:5])
```

```python
#Slice from the end with Negative indexing

str = "Hello, World!"
print(str[2:-3])
```

```python
# Slice usign this Syntex : string[start:stop:step]

str = "Hello, "
str1 = "Aman"
str2 = str+str1
print(str2[::2])
```

#Q2. Explain the key features of lists in Python.

####List is a built-in data structure that allows multiple items to be stored in a single variable.
>(a) Lists are:

>>Ordered - They maintain the order of elements.

>>Mutable - Items can be changed after creation.
>>Allow duplicates - They can contain duplicate values

>(b) Access List Elements- Each element in a list is associated with a number, known as index.The index always starts from 0. The first element of a list is at index 0, the second element is at index 1, and so on.

>(c) Adding Elements- append() method to add an element to he end of a Python list.

>(d) Change List Items- The items of a list by assigning new values using the indexes and = (Assignment Operator).

>(e) Remove Item from the list- We can remove an item from a list using remove() method.

>(f) List Length- len() function is used find the length of elements in a list.


```python
# Access Element Using Index

list = ["apple", "banana", "cherry"]
print(list[1])
print(list[2])
```

```python
# Adding Elements

list = ["apple", "banana", "cherry"]
list.append("orange")
print(list)
```

```python
# Change List Items

list = ["apple", "banana", "cherry"]
list[1] = "blackcurrant"
print(list)
```

```python
# Remove items from a list

list = ["apple", "banana", "cherry"]
list.remove("banana")
print(list)
```

```python
# Length of a List

list = ["apple", "banana", "cherry"]
print(len(list))
```

```python
# Shallow Copy

list1 = ["apple", "banana", "cherry"]
list2 = list1
print(list2)
```

#Q5. Describe the key features of sets and provide examples of their use.

###A set is a mutable, unordered group of elements, where the elements themselves are immutable.

>Sets are created by using curly braces{}.

>The elements in Python sets must have an immutable data type. Mutable data types, like lists and dictionaries, cannot be added to sets.

>Python sets are unordered

>Add Items to a Set: add() method is used add am item to a set.

>Update Items to a Set: Update() method is used to update the set with items other collection types.

>Remove an Element from a Set: The discard() method to remove the specified element from a set.

>Clear all the Element from the Set: The clear() method is used to clear all the elements of the set.

>To get the maximum element from the set: The max() function is used to get the maximum element from the set.

>To get the minimum element from the set: The min() function is used to get the minumum element from the set.

>Two sets A and B:
>>The Operator '|' is used to perform the set union operation.

>>The Operator '&' is used to perform the set intersection operation.

>>The Operator '-' is used to perform the set Difference b/w two sets.

>>The Operator '^' is used to perform the syemmetric differences b/w two sets.

```python
#Creating a set
set1 = {"apple", "banana", "cherry"}
print(set1)
```

```python
#Sets are Immutable
set1 = {"apple", "banana", "cherry"}
#To add any element we have to use add() method
set1.add("orange")
print(set1)
```

```python
#update Method
set1 = {"apple", "banana", "cherry"}
set1.update(["orange", "mango", "grapes"])
print(set1)
```

```python
#Discard method
set1 = {"apple", "banana", "cherry"}
set1.discard("banana")
print(set1)
```

```python
#Clear method
set1 = {"apple", "banana", "cherry"}
set1.clear()
print(set1)
```

```python
#min and max methods
set1 = {"apple", "banana","strawberry", "cherry"}
print(min(set1))
print(max(set1))
```

```python
#set Operation
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)
print(set1 ^ set2)
```

#Q7. Describe how to add, modify, and delete items in a dictionary with examples.


####A dictionary in Python is an unordered collection of key-value pairs. It is a mutable data type, which means that we can modify its contents by adding, changing, or removing elements

>Add Items:- To add a new key-value pair to a dictionary, we can use the assignment operator '='.

>>We can also add multiple key-value pairs at once using the update() method

>Modify Items:-To change the value associated with a particular key in a dictionary, we can simply reassign the value using the assignment operator =.

>Remove Items:
>> By using del Statement:-
We can use the del statement to remove a specific key-value pair from a dictionary

>>By using pop() method:-We can use the pop() method to remove a specific key-value pair from a dictionary and return its value

>Update() method:- We can use the update() method to add the key-value pairs from one dictionary to another.

```python
#Creating a Dictionary
dict1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
  }

print(dict1)
```

```python
#adding items to the dictionary
dict1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
  }

dict1["Origin"] = "India"
print(dict1)
```

```python
#adding element using update method
dict1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
  }

dict1.update({"Origin": "India"})
print(dict1)
```

```python
#remove items using del method
dict1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
  }

del dict1["age"]
print(dict1)
```

```python
#remove itmes usign pop method
dict1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
  }

dict1.pop("age")
print(dict1)
```

#Q8. Discuss the importance of dictionary keys being immutable and provide examples.


###dictionary keys being immutable:-

>Hashability and Dictionary Keys:
>>1. Dictionary keys must be immutable because dictionaries use a hash table implementation. The hash value of a key is calculated to find the corresponding value.

>>2. If a key were mutable (i.e., its value could change), its hash would also change. This would lead to inconsistencies when looking up values in the dictionary.

>>3. By using immutable keys, Python ensures that the hash remains constant throughout the program’s execution.

>Benefits:
>>1. Immutable keys ensure consistency in dictionary lookups.

>>2. They allow efficient hash-based indexing.

>>3. Tuples, being hashable, can serve as keys, unlike mutable lists.



```python
# Dictionary keys must be immutable
dict1 = {
  [1, 2, 3]: "John",
  "age": 36,
  "country": "Norway"
  }
dict1
```

```python
dict1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
  }

dict1["name"] = "John Doe"
print(dict1)
```

```python
my_dict = {
    ('John', 'Doe'): 28,
    ('Jane', 'Smith'): 32
}
print(my_dict)
```
